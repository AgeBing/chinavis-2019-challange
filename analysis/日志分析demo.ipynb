```python
import pandas as pd
import json
```

```python
files = [
    'data/day1.csv',
    'data/day2.csv',
    'data/day3.csv'
]

def sec2hour(sec):
    hour = int(sec / 3600)
    minite = int(sec / 60)  - hour * 60
    second = sec - hour * 3600 - minite*60
    print(str(hour) + ':' + str(minite) + ':' + str(second) )
```

```python
df_day1 = pd.read_csv(files[0])
df_day2 = pd.read_csv(files[1])
df_day3 = pd.read_csv(files[2])
```

## 三天会议的时间分布

```python
sec2hour(df_day1.time.min())
sec2hour(df_day1.time.max())
sec2hour(df_day2.time.min())
sec2hour(df_day2.time.max())
sec2hour(df_day3.time.min())
sec2hour(df_day3.time.max())
```
